# Análise de dados de filmes exebidos no brasil de 2009 a 2017
Com base nos dados da Ancine (http://oca.ancine.gov.br) Vamos analisar of filmes exibidos no Brasil
Arquivo -> 2120_0.xlsx

```python
# Importando as bibliotecas
import pandas as pd
# Configurando o matplotlib
%matplotlib inline
```

```python
# Criando o DataFrame com base no arquivo exel
# Precisamos pular as duas primeiras linhas (por causa do titulo do arquivo) e as 16 ultimas (lixo)
df = pd.read_excel('2120_0.xlsx', skiprows=2, skipfooter=16)

# Exibindo as 5 primeiras linhas do DataFrame
df.head()
```

```python
#Exibindo as 5 ultimas
df.tail(5)
```

```python
# Verificar todos os tipos das colunas
df.dtypes
```

```python
# Converter o campo Data de lançamento em data
df['Data de lançamento'] = pd.to_datetime(df['Data de lançamento'])
```

```python
# Vamos tentar identificar o problema da coluna Data de lançamento
df['Data de lançamento'].value_counts()
```

```python
# Exibir alguns filmes que estao com data relançamento
df[df['Data de lançamento'] == 'Relançamento'].head()
```
